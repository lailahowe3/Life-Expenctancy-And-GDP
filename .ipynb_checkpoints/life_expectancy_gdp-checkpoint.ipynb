# Life Expectancy and GDP
The main goal of this work is to determine the relationship if any between the life expectancy and the GDP of 6 different countries. The data will come from the World Health Oraganisation and the World Bank and is stored in our 'all_data' CSV file.

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
# load and view data
all_data = pd.read_csv('all_data.csv')
print(all_data.head())
```

```python
#check for any missing data values
print(all_data.info())
```

```python
# look at the different countries that are present in the data
print(all_data['Country'].unique())
```

```python
# look at the range of years we have available
print(all_data.Year.min())
print(all_data.Year.max())
```

```python
# find the summary statistics for life expectancy per country
print(all_data[['Country','Life expectancy at birth (years)']].groupby('Country').describe())
```

From the summary statistics displayed above we can see that the mean, min and max values for life expectancy at birth for Zimbabwe are less than the values for the other countries. Zimbabwe also has the largest standard deviation. The highest mean is for Germany which also has the highest min and max values.

```python
# find the summary statistics for GDP per country
print(all_data[['GDP','Country']].groupby('Country').describe())
```

Once again from looking at the summary statistics described above we can see that Zimbabwe has the lowest mean, min and max for GDP. The USA has the highest mean, min and max GDP.

Plotting graphs to show the relationships between life expectancy and GDP over time per country as well as the relationship for life expectancy and GDP per country will more easily demostrate any relationships.

```python
# save each countries data to a new df
data_chile = all_data[all_data.Country == 'Chile']
data_china = all_data[all_data.Country == 'China']
data_germany = all_data[all_data.Country == 'Germany']
data_mexico = all_data[all_data.Country == 'Mexico']
data_usa = all_data[all_data.Country == 'United States of America']
data_zimbabwe = all_data[all_data.Country == 'Zimbabwe']
# plot a 6 lines to show the relationship between life expectancy and year for each country.
plt.figure(figsize=(10,8))
plt.plot(data_chile.Year, data_chile['Life expectancy at birth (years)'],label='Chile')
plt.plot(data_china.Year,data_china['Life expectancy at birth (years)'],label='China')
plt.plot(data_germany.Year,data_germany['Life expectancy at birth (years)'],label='Germany')
plt.plot(data_mexico.Year,data_mexico['Life expectancy at birth (years)'],label='Mexico')
plt.plot(data_usa.Year,data_usa['Life expectancy at birth (years)'],label='USA')
plt.plot(data_zimbabwe.Year,data_zimbabwe['Life expectancy at birth (years)'],label='Zimbabwe')
plt.xlabel('Years')
plt.ylabel('Life expectancy at birth (years)')
plt.title('Life expectancy at birth vs Year')
plt.legend()
plt.show()
plt.clf()
```

Findings: While there looks to be a general trend that as time increases life expectancy increases it is difficult to tell exactly the trends in the data as Zimbabwe life expectancy is much lower than the other countries. To investigate this further these will be plotted seperately to see the trends for each country. We can tell that Zimbabwe has the lowest life expectancy and Germany has the highest throughout the years.

```python
# plot individual graphs for each country for Life expectancy vs Year.
plt.figure(figsize=(20,16))
plt.subplot(2,3,1)
plt.plot(data_chile.Year, data_chile['Life expectancy at birth (years)'],label='Chile',color='blue')
plt.xlabel('Years')
plt.ylabel('Life expectancy at birth (years)')
plt.title('Chile Life expectancy at birth (years) vs Year')
plt.subplot(2,3,2)
plt.plot(data_china.Year,data_china['Life expectancy at birth (years)'],label='China',color='orange')
plt.xlabel('Years')
plt.ylabel('Life expectancy at birth (years)')
plt.title('China Life expectancy at birth (years) vs Year')
plt.subplot(2,3,3)
plt.plot(data_germany.Year,data_germany['Life expectancy at birth (years)'],label='Germany',color='green')
plt.xlabel('Years')
plt.ylabel('Life expectancy at birth (years)')
plt.title('Germany Life expectancy at birth (years) vs Year')
plt.subplot(2,3,4)
plt.plot(data_mexico.Year,data_mexico['Life expectancy at birth (years)'],label='Mexico',color='red')
plt.xlabel('Years')
plt.ylabel('Life expectancy at birth (years)')
plt.title('Mexico Life expectancy at birth (years) vs Year')
plt.subplot(2,3,5)
plt.plot(data_usa.Year,data_usa['Life expectancy at birth (years)'],label='USA',color='purple')
plt.xlabel('Years')
plt.ylabel('Life expectancy at birth (years)')
plt.title('USA Life expectancy at birth (years) vs Year')
plt.subplot(2,3,6)
plt.plot(data_zimbabwe.Year,data_zimbabwe['Life expectancy at birth (years)'],label='Zimbabwe',color='brown')
plt.xlabel('Years')
plt.ylabel('Life expectancy at birth (years)')
plt.title('Zimbabwe Life expectancy at birth (years) vs Year')
plt.show()
plt.clf()
```

Findings: We can see there is a relationship between time and life expectancy. For all 6 countries as time goes on there is a general increase in life expectancy. Zimbabwe has a dip in life expectancy around 2004 and Mexico has a dip around 2008-2011. 

```python
#plot 6 lines to show the relationship between GDP and time per country.
plt.figure(figsize=(10,8))
plt.plot(data_chile.Year, data_chile['GDP'],label='Chile')
plt.plot(data_china.Year,data_china['GDP'],label='China')
plt.plot(data_germany.Year,data_germany['GDP'],label='Germany')
plt.plot(data_mexico.Year,data_mexico['GDP'],label='Mexico')
plt.plot(data_usa.Year,data_usa['GDP'],label='USA')
plt.plot(data_zimbabwe.Year,data_zimbabwe['GDP'],label='Zimbabwe')
plt.xlabel('Years')
plt.ylabel('GDP')
plt.title('GDP vs Year')
plt.legend()
plt.show()
plt.clf()
```

Findings: As the USA has such a large GDP it is hard to tell what the trend is for Zimbabwe and Chile. For these it might be easier to plot on a seperate graph so that we can see their relationships between time and GDP more clearly. However, this graph does show a large increase in GDP from 2000 to 2015 for the USA and China.

```python
# plot individual graphs for each country for GDP vs Year.
plt.figure(figsize=(20,16))
plt.subplot(2,3,1)
plt.plot(data_chile.Year, data_chile['GDP'],label='Chile',color='blue')
plt.xlabel('Years')
plt.ylabel('GDP')
plt.title('Chile GDP vs Year')
plt.subplot(2,3,2)
plt.plot(data_china.Year,data_china['GDP'],label='China',color='orange')
plt.xlabel('Years')
plt.ylabel('GDP')
plt.title('China GDP vs Year')
plt.subplot(2,3,3)
plt.plot(data_germany.Year,data_germany['GDP'],label='Germany',color='green')
plt.xlabel('Years')
plt.ylabel('GDP')
plt.title('Germany GDP vs Year')
plt.subplot(2,3,4)
plt.plot(data_mexico.Year,data_mexico['GDP'],label='Mexico',color='red')
plt.xlabel('Years')
plt.ylabel('GDP')
plt.title('Mexico GDP vs Year')
plt.subplot(2,3,5)
plt.plot(data_usa.Year,data_usa['GDP'],label='USA',color='purple')
plt.xlabel('Years')
plt.ylabel('GDP')
plt.title('USA GDP vs Year')
plt.subplot(2,3,6)
plt.plot(data_zimbabwe.Year,data_zimbabwe['GDP'],label='Zimbabwe',color='brown')
plt.xlabel('Years')
plt.ylabel('GDP')
plt.title('Zimbabwe GDP vs Year')
plt.show()
plt.clf()
```

These figures show more easily the trends in GDP for each country. From our original plot we could not tell there was a decrease in GDP from 2000 to 2008 in Zimbabwe. We can see that across all 6 countries there is an increase in GDP with time.

Now that we have seen a general trend for each country that as the years have passed both Life expectancy and GDP have increased from 2000 to 2015. We can now examine the relationship between Life expectancy and GDP for each country.

```python
# plot individual graphs for each country for GDP vs Life expectancy.
plt.figure(figsize=(20,16))
plt.subplot(2,3,1)
plt.scatter(data_chile['Life expectancy at birth (years)'], data_chile['GDP'],label='Chile',color='blue')
#sns.lmplot(x='Life expectancy at birth (years)', y='GDP', data = data_chile, line_kws={'color': 'black'}, lowess=True)
plt.xlabel('Life expectancy at birth (years)')
plt.ylabel('GDP')
plt.title('Chile GDP vs Life expectancy at birth (years)')
plt.subplot(2,3,2)
plt.scatter(data_china['Life expectancy at birth (years)'],data_china['GDP'],label='China',color='orange')
#sns.lmplot(x='Life expectancy at birth (years)', y='GDP', data = data_china, line_kws={'color': 'black'}, lowess=True)
plt.xlabel('Life expectancy at birth (years)')
plt.ylabel('GDP')
plt.title('China GDP vs Life expectancy at birth (years)')
plt.subplot(2,3,3)
plt.scatter(data_germany['Life expectancy at birth (years)'],data_germany['GDP'],label='Germany',color='green')
#sns.lmplot(x='Life expectancy at birth (years)', y='GDP', data = data_germany, line_kws={'color': 'black'}, lowess=True)
plt.xlabel('Life expectancy at birth (years)')
plt.ylabel('GDP')
plt.title('Germany GDP vs Life expectancy at birth (years)')
plt.subplot(2,3,4)
plt.scatter(data_mexico['Life expectancy at birth (years)'],data_mexico['GDP'],label='Mexico',color='red')
#sns.lmplot(x='Life expectancy at birth (years)', y='GDP', data = data_mexico, line_kws={'color': 'black'}, lowess=True)
plt.xlabel('Life expectancy at birth (years)')
plt.ylabel('GDP')
plt.title('Mexico GDP vs Life expectancy at birth (years)')
plt.subplot(2,3,5)
plt.scatter(data_usa['Life expectancy at birth (years)'],data_usa['GDP'],label='USA',color='purple')
#sns.lmplot(x='Life expectancy at birth (years)', y='GDP', data = data_usa, line_kws={'color': 'black'}, lowess=True)
plt.xlabel('Life expectancy at birth (years)')
plt.ylabel('GDP')
plt.title('USA GDP vs Life expectancy at birth (years)')
plt.subplot(2,3,6)
plt.scatter(data_zimbabwe['Life expectancy at birth (years)'],data_zimbabwe['GDP'],label='Zimbabwe',color='brown')
#sns.lmplot(x='Life expectancy at birth (years)', y='GDP', data = data_zimbabwe, line_kws={'color': 'black'}, lowess=True)
plt.xlabel('Life expectancy at birth (years)')
plt.ylabel('GDP')
plt.title('Zimbabwe GDP vs Life expectancy at birth (years)')
plt.show()
plt.clf()
```

For all 6 countries we see that with an increase in the life expectancy there is also an increase in GDP. For the USA the relationship looks linear but for China the relationship looks exponential. Either way it is clear that there is a positive correlation between life expectancy and GDP for each of our 6 countries.

# Conclusion:
In conclusion we can see that between 2000 and 2015 there has been an increase in GDP and Life expectancy for all 6 countries data we have examined. We can also see that as Life expectancy increases GDP also increases for each country. There is a positive correlation between Life expectancy and GDP.
